```python
# 3.6.c
def replace_nan_horsepower(df):
    # Calculate mean horsepower per cylinder group
    horsepower_means = df.groupby('cylinders')['horsepower'].mean()

    # Loop through rows with NaN and replace with the mean
    for idx, row in df[df['horsepower'].isna()].iterrows():
        cylinder_group = row['cylinders']
        df.at[idx, 'horsepower'] = horsepower_means[cylinder_group]

    return df
```
